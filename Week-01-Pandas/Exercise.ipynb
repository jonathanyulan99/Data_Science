## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 

## price_greater_than_100 is the variable name of the condition we are going to search
## the . operator can be used to access specific columns of the df we are working with 
## here we can use the .shape[] method that can return certain manipulations of the df such as the #of rows
## we can also use the len() operator to obtain all the rows or the 'length' of the data frame which is variable 
price_greaterthan100 = df['price'] < 100
data_frame_output = df[price_greaterthan100]
print("The Number of Listings with Price Greater Than 100:" , len(data_frame_output))
## print out the data frame with a method called .iloc[beginning_index : ending_index] will go up to the end unlike range()
data_frame_output.iloc[0:len(data_frame_output)]
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

## brooklyn_listings_only will be the name of the condition we use
brooklyn_listings_only = df.neighbourhood_group == "Brooklyn"
## using df_bk as the name of the new data frame we create one with the condition we defined in the previous line
## df_bk is a variable name of a data_frame that is created using the df initialized in the beginning with the condition as the 
## ... [column_condition]
df_bk = df[brooklyn_listings_only]
## using the python built-in condition length() we can find the total size of the df_bk that we created 
print("The Number of Listings in Brooklyn are: ", len(df_bk))
df_bk.head()
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.

## Step1: Create condition to use for the first part of this problem
brooklyn_condition = df.neighbourhood_group == 'Brooklyn'
## Step2: Create second condition to use for the second part of this problem
price_condition = df.price < 100
## Step3: create the data_frame_ouput using a user-defined variable to assign to both conditions
## ** and => & or => | ** CANNOT USE WORD EQUIVALENT UNLIKE PYTHON this is in PANDAS library
df_brook_price = df[brooklyn_condition & price_condition]
## Step4: print out the number of rows using either len(df_object) or the df_object.shape[0] to only give the number of rows/
print("The Number of Listings in Brooklyn that are priced less than 100: ", len(df_brook_price))
df_brook_price
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
## Step1: Create New Data_Frame Variable Object 
condition_michael_david_john_daniel = df['host_name'].isin(['Michael','David','John','Daniel'])
## Step2: use that new condition to create a new Data Frame
## df[new_condition] to join from the original data frame object we created initially
df[condition_michael_david_john_daniel]
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

##Step 1: Create the first condition 
williamsburg_condition = df['neighbourhood'] == 'Williamsburg'
## Step2: Create the new data frame column with the syntax:
## data_frame_object['new_col_name'] = df['old_column_name']
df['adjusted_price'] = df['price']
## Step3: use the np.where() to create a condition that adds +100 to the price
## np.where(condition_user_defined,condition_applies_change,condition_doesn't_apply_not_changed)
df['adjusted_price'] = np.where(williamsburg_condition,df.adjusted_price+100,df.adjusted_price)

## STEP4:print out the the data_frame and see the new adjusted_price column added
df[['neighbourhood','price','adjusted_price']].loc[50:1000]
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  
## lets get the total number of entries or rows using either len() or .shape[0]
total_rooms = df.shape[0]

## Get the private room condition using len alternatively could use .shape[0]
private_condition = len(df[df['room_type'] == 'Private room'])

## Print the private room percent 
print(f'Private Rooms(Percentage): {private_condition*100/total_rooms} %')

## Do the same process for the shared room condition 
shared_condition = df[df['room_type']=='Shared room'].shape[0]
print(f'Shared Rooms(Percentage): {shared_condition*100/total_rooms}%')

## There is also a .value_counts method that is similiar 
## alternatively:

### .counts() iterations of the value 
### .value() of all the types of room_type 
### this would be 2 because of the room type: shared and private 
 ##   data_frame_percent = df['room_type'].value.counts()/df['room_type'].counts().sum()*100
 ##   data_frame_percent['Private room'], data_frame_percent['Shared room']
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

##Step1: define a grouping_name_variable and then use pandas.groupby function
##Step2: groupby(by='specific_column')
g_neighbourhood_group = df.groupby(by='neighbourhood_group')
print('Number of Listings in Distinct Neighborhood Groups: ')
##Step3:The grouping_name_variable is printed using the .size() method
g_neighbourhood_group.size()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

##Step1: We are using the g_neighbourhood_group variable that represents a data frame 
##Step2: Using this variable_name that has all the listings of all the neighborhood_groups
##Step3: Using the same df['column_attribute'].function()
## .function being .mean() that works on the df object called the neighborhood_group_variable
g_neighbourhood_group['price'].mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

## Rememeber we already have the groupby above for the specific neighborhood groups
## Use that same .groupby data frame user defined variable and apply the .agg
## .agg(['list of functions', no_comma_if_user_defined_function])
g_neighbourhood_group['price'].agg(['min','max'])
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.

##Create new groupby variable name
##Here we are grouping by two column names, thus we need to create a list
## ['column_names_in_df','...',...]
## new variable name data frame defined can be applied with pandas standard library of functions 
## df_object.mean()
g_neighbourhood_group_room_type = df.groupby(['neighbourhood_group', 'room_type'])
g_neighbourhood_group_room_type['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

## Here it is once again asking for specific panda defined fucntions in a specific grouping of the df
## that df grouping is based on the room type and neighbourhood_group 
## df.groupBy creates the new talble based on the list elements in the function .groupBy(['columns'])
## df.agg applies the standard functions in pandas using a string format
new_grouping = df.groupby(['neighbourhood_group','room_type']).agg(['count','min','max','mean','median','std'])
## prints the new data frame based on the condition of price that is not represented on the data frame as column
## but rather as the characterisstic being worked on in the table 
new_grouping['price']
```
